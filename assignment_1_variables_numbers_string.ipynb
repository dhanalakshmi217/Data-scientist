```

Total Sales for the Day: $890.0
```

````python
# write your code here
product1_price = 20.0  # Price per unit for product 1
product1_qty = 15  # Quantity sold of product 1

product2_price = 35.0  # Price per unit for product 2
product2_qty = 10  # Quantity sold of product 2

product3_price = 12.0  # Price per unit for product 3
product3_qty = 20
total1 = product1_qty * product1_price
total2 = product2_qty * product2_price
total3 = product3_qty * product3_price
daily_total = total1 + total2 + total3
print("Daily Sales Summary:")
print(f"- Product 1: Sold {product1_qty} units at ${product1_price} each, Total: ${total1}")
print(f"- Product 2: Sold {product2_qty} units at ${product2_price} each, Total: ${total2}")
print(f"- Product 3: Sold {product3_qty} units at ${product3_price} each, Total: ${total3}")
print(f"Total Sales for the Day: ${daily_total}")
````
