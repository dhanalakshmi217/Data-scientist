```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
np.random.seed(42)
data = {
    "Math_Score":np.random.randint(40,100,30),
    "Science_Score":np.random.randint(35,95,30)
}
df=pd.DataFrame(data)
print("Dataset:\n",df)
```

```python
des_stats=df.describe()
print("Descriptive statistics:\n",des_stats)
```

```python
central_tendency={"Mean":df.mean(),"Median":df.median(),"Mode":df.mode().iloc[0]}
print("\n Central_Tendency:\n",pd.DataFrame(central_tendency))
```

```python
dispersion={"Variance":df.var(),"Standard Deivation":df.std(),"Range":df.max()-df.min(),"IQR":df.quantile(0.75) - df.quantile(0.25)}
print("\n Dispersion Measures:\n",pd.DataFrame(dispersion))
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Science_Score"],bins=6,kde=True,color='skyblue',edgecolor='black')
plt.title("Histogram of Math Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Math_Score"],color='purple')
plt.title("Box plot of Science Scores")
plt.xlabel("Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```
